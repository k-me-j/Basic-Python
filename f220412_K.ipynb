```python
from matplotlib import pyplot as plt
import math
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
X = [(i * math.pi / 180) for i in range(0, 361, 30)]
Y = [math.sin(i * math.pi / 180) for i in range(0, 361, 30)]
```

```python
X, Y
```

```python
plt.scatter(X, Y)
plt.show()
```

```python
X_2d = [[i] for i in X]   # 2차원이어야 fit 가능해서 2차원 만들어주기
X_2d
```

```python
lrg = LinearRegression()
lrg.fit(X_2d, Y)
```

```python
plt.scatter(X, Y)
plt.plot(X, lrg.predict(X_2d))  # 잔차가 최소가 되는 직선 -> 선형으로 풀 수 없는 문제라서 그래
plt.show()
```

# numpy로 만들기

```python
X2 = np.linspace(0, 2*math.pi, 12)  # 0 ~ 2*파이까지 12개 선형으로 생성
Y2 = np.sin(X2)
X2
```

```python
plt.scatter(X2, Y2)
plt.show()
```

```python
X2_2d = X2.reshape(-1,1)  # np를 reshape할 때 -1을 주면 차원이 하나 늘어남
X2_2d
```

```python
lrg = LinearRegression()
lrg.fit(X2_2d, Y2)
```

```python
plt.scatter(X2, Y2)
plt.plot(X2, lrg.predict(X2_2d))  # 잔차가 최소가 되는 직선 -> 선형으로 풀 수 없는 문제라서 그래
plt.show()
```

```python
n = 100

x = np.random.rand(n, 1) * 6 - 3  # -3 ~ 3 사이로 맞춰주기 위해 *6-3 들어감
# x
```

```python
y = 0.5 * x**2 + x + 2 + np.random.rand(n, 1)
# y
```

```python
plt.scatter(x, y, s=3)
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_feat = PolynomialFeatures(degree=2)
```

```python
x_poly = poly_feat.fit_transform(x)
# x_poly
```

```python
lrg = LinearRegression()
lrg.fit(x_poly, y)
```

```python
plt.scatter(x, lrg.predict(x_poly), s=3)
plt.show()
```
