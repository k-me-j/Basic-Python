````python
import numpy as np
np.random.seed(8)


def gen_data(mat, n, m):
    '''
    mat = n x m matrix 생성
    '''
    for i in range(n):
        for j in range(m):
            mat[i][j] = np.random.randint(0, 10)
        
    
def matmul(mat1, mat2, result):
    '''
    element-wise가 아닌 matrix mulitplication 함수
    n x m * m x n = n x n 반환
    '''
    for i in range(len(result)):
        for j in range(len(result[i])):
            for k in range(len(mat1[i])):
                result[i][j] += mat1[i][k] * mat2[k][j]
        
        
if __name__ == "__main__":    
    n = np.random.randint(3, 20)
    m = np.random.randint(3, 20)
    
    print(n, m)
    
    mat1 = np.zeros((n, m))
    mat2 = np.zeros((m, n))
    result = np.zeros((n, n))
    
    gen_data(mat1, n, m)
    gen_data(mat2, m, n)
    matmul(mat1, mat2, result)
    
    # print(mat1)
    # print(mat2)
    print(result, result.shape)
````

````python
import numpy as np

def inner_product(vectors, in_vec):
    '''
    vectors 중 in_vec와 가장 유사한 row를 (인덱스, 값) 형태로 반환
    '''
    max_similar = (0, 0)  # (idx, [round(val, 3) for val in row])
    
    for i in range(len(vectors)):
        inner_p = np.inner(vectors[i], in_vec)
        if inner_p > max_similar[1]:
            max_similar = (i, round(inner_p, 2))
            # print(vectors[i])
        
    return max_similar
    
    
if __name__ == "__main__":
    np.random.seed(1234)
    
    doc_vec = np.zeros((100, 6))
    for i in range(100):
        for j in range(6):
            doc_vec[i][j] = np.random.random()
    
    # print(doc_vec, doc_vec.shape)
    
    vec_doc_1 = [0.2, 0.4, 0.02, 0.1, 0.9, 0.8]
    
    print(inner_product(doc_vec, vec_doc_1))
````
