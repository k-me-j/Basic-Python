````python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

import numpy as np

from file_read import read_value
````

````python
final_points, cnt = read_value("score_mlr03.txt")
````

## MxN Matrix 형태로 만들기

````python
A = np.array(final_points)
A = A[:-5, :-1]
A = A.reshape(cnt-5, 3)  # train_data에 맞춘 크기 (20x3)

b = np.array(final_points)
b = b[:-5, -1]
b = b.reshape(cnt-5, 1)  # test_data에 맞춘 크기 (20x1)
````

## x_hat : 학습 기준으로는 weight값과 동일

````python
x_hat1 = np.matmul(A.T, A)
x_hat2 = np.linalg.inv(x_hat1)  # x_hat1의 역행렬 구하기
x_hat3 = np.matmul(x_hat2, A.T)
x_hat = np.matmul(x_hat3, b)
x_hat
````

````python
test_A = np.array(final_points)
test_A = test_A[-5:, :-1]

pred = np.matmul(test_A, x_hat)
pred
````

## weight, bias의 초기값은 랜덤 값으로 셋팅됨
- model().parameter -> random값이 들어가있음
- iteration or epoch 돌면서 model parameter의 값들을 업데이트
- 실행마다 결과값이 다른 것을 방지하기 위해
```torch.manual_seed(1)``` 사용 가능

````python
torch.manual_seed(1)
````

train data를 일단 가져오고, train data로부터 x,y 값을 구분해야 함
- x = [mid, final, report]
- y = [final points]

````python
train_data = np.array(final_points)
train_data.shape
````

````python
x_train = train_data[:-5, :-1]  # 20x3 생성
y_train = train_data[:-5, -1]
````

````python
x_test = np.array(final_points)
x_test = x_test[-5:, :-1]
````

````python
x_train
````

````python
y_train
````

# pytorch 입력은 Tensor이므로 바꿔주기

````python
x_train = torch.FloatTensor(x_train)
y_train = torch.FloatTensor(y_train)

x_test = torch.FloatTensor(x_test)
````

# 모델 선언, 가중치 초기화
```nn.Linear(input dimension, output dimension)```

````python
model = nn.Linear(3, 1)  # column 3개, 출력 1개

# optimizer setting
optimizer = optim.SGD(model.parameters(), lr=0.00006)    # 0.000005
# optimizer = optim.Adam(model.parameters(), lr=0.000005)  
````

# training

````python
nb_epochs = 1200000

for epoch in range(nb_epochs + 1):
    # forwarding ( H(x) 계산 : wx + b )
    pred = model(x_train)  # predict

    cost = F.mse_loss(pred, y_train)  # cost가 최소화 되도록 학습

    optimizer.zero_grad()  # 누적 방지

    cost.backward()  # backward 연산
    optimizer.step()  # weight, bias update

    if epoch % 100 == 0:
        print(f"Epoch : {epoch:4d}, Cost : {cost.item():.4f}")
````

````python
model_params = list(model.parameters())  # model weight 값들 들어있음 : weight 3개 + bias 1개
model_params[0], model_params[1]
````
