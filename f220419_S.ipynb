# ▶ 시도별 전출입 인구수

````python
df = pd.read_excel('../dataset/학습데이터처리_Data/시도별 전출입 인구수.xlsx')
df.head(3)
````

````python
df.info()
````

## ```df.fillna()``` 기본 사용법
```python
df.fillna(value=None, method=None, axis=None, inplace=False, limit=None, downcast=None)
```
- value : 결측값을 대체할 값입니다. dict형태로도 가능합니다.
- method : 결측값을 변경할 방식입니다. bfill로 할경우 결측값을 바로 아래 값과 동일하게 변경합니다.
    - ffill로 할 경우 결측값을 바로 위 값과 동일하게 변경합니다.
- axis : {0 : index / 1 : columns} fillna 메서드를 적용할 레이블입니다.
- inplace : 원본을 변경할지 여부입니다. True일 경우 원본을 변경하게 됩니다.
- limit : 결측값을 변경할 횟수입니다. 위에서부터 limit로 지정된 갯수만큼만 변경합니다.
- downcast : 다운캐스트할지 여부입니다. downcast='infer'일 경우 float64를 int64로 변경합니다.

````python
# '전출지별' 피처는 똑같은 값들이 반복되지 않고 첫 값 이후로는 null로 지정되어 있음. 그걸 채워주기 위한 작업
df1 = df.fillna(method='ffill')
df1.head(30)
````

````python
# 서울 -> 타지로 이동한 인원 확인
mask = (df1['전출지별'] == '서울특별시') & (df1['전입지별'] != '서울특별시')
mask[19:19+18]  # 전출지가 서울특별시인 것만 슬라이싱해서 확인해보기
````

````python
df_seoul = df1[mask]
df_seoul.head()
````

````python
# df_seoul는 전출지가 서울특별시인 것들만 모아둔 것이기 때문에 전출지별 피처 지우기
df_seoul.drop(['전출지별'], axis=1, inplace=True)
````

````python
df_seoul.rename({'전입지별':'전입지'}, axis=1, inplace=True)  # 열 이름 변경
df_seoul.set_index('전입지', inplace=True)  # 인덱스 설정
df_seoul.head()
````

````python
sr = df_seoul.loc['경기도']
sr.head()
````

````python
plt.plot(sr)
plt.show()
````

````python
plt.plot(sr)
plt.title("서울->경기도 인구 이동")
plt.xlabel("기간(연도순)")
plt.ylabel("이동 인구수")
plt.show()
````

# ▶ 과제
1. 서울 -> 대구로 이동한 인구수에 대한 선 그래프 그려보기
2. mask 형태로도 한 번 실습해보고, df.iloc 이용해서도 실습해보기
- dash : 이동인구가 없단 의미. ```df.replce('-', 0, inplace=True)``` 로 인구 이동 0명 숫자로 변경해주기

## ▷ 1. loc로 가져오기

````python
dg = df_seoul.loc['대구광역시']
dg.head(15)
````

````python
dg.replace('-', 0, inplace=True)
````

````python
plt.plot(dg)
plt.title("서울->대구광역시 인구 이동")
plt.xlabel("기간(연도순)")
plt.ylabel("이동 인구수")
plt.xticks(rotation='vertical')   # x축 이름 세로로 세우기
plt.rcParams["figure.figsize"] = (10, 5)  # figure size 조절
plt.show()
````

## ▷ 2. mask 사용해보기

````python
mask_dg = (df1['전출지별'] == '서울특별시') & (df1['전입지별'] == '대구광역시')
mask_dg
````

````python
df_daegu = df1[mask_dg]
df_daegu.head()
````

````python
df_daegu.drop(['전출지별'], axis=1, inplace=True)  # 어차피 서울 출발인 데이터들만 있으니 전출지 필요없어서 삭제
df_daegu.rename({'전입지별':'전입지'}, axis=1, inplace=True)  # 열 이름 변경. 어차피 대구만 있으니까
df_daegu.set_index('전입지', inplace=True)  # 인덱스를 '전입지'로 설정
df_daegu.replace('-', 0, inplace=True)  # -로 표현된 0명을 숫자로 변경해주기
df_daegu.head()
````

````python
df_daegu = df_daegu.T  # T 안해주면 x 축에 '대구광역시'만 뜸
````

````python
plt.plot(df_daegu)
plt.title("서울->대구광역시 인구 이동")
plt.xlabel("기간(연도순)")
plt.ylabel("이동 인구수")
plt.xticks(rotation='vertical')   # x축 이름 세로로 세우기
plt.rcParams["figure.figsize"] = (10, 5)  # figure size 조절
plt.show()
````
