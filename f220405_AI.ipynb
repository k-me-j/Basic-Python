# 인자로 함수를 받음
: C에선 function point라고 한다나봐

```python
def showGraph(function, arg):
    x = arg
    y = function(x)
    plt.plot(x, y)
    plt.show()
```

```python
showGraph(np.sin, c)
```

```python
showGraph(np.sin, d)
```

```python
showGraph(np.cos, c)
```

```python
showGraph(np.tan, c)
```
